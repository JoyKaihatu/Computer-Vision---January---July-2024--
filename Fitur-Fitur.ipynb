````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
````

````python
path = 'Shape1.png'

img = cv2.imread(path)

plt.axis('off')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img, cmap='gray')


print(img.shape)
````

# Harris Corner Detection

Tipe Deteksi : Corner

Syntax: 

`cv2.cornerHarris(image, dest, blockSize, kSize, k)`

Parameters:  

- Image – Gambar yang akan dipakai
- Dest – Variabel yang digunakan untuk menyimpan image
- Block size – Neighborhood size
- Ksize – Free Parameter Variabel for Sobel Operator
- k - Free parameter for harris corner detection




````python
image = np.copy(img)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
gray_image = np.float32(gray_image) 
  
 
dst = cv2.cornerHarris(gray_image, blockSize=2, ksize=3, k=0.04) 
  
 
dst = cv2.dilate(dst, None) 
image[dst > 0.01 * dst.max()] = [0, 255, 0] 



  
plt.imshow(image) 
plt.show()
````

# Shi-Tomasi Corner Detection

Syntax: 

```cv2.goodFeaturesToTrack(image, maxc, Quality, maxD)```

Parameters:

- image – Gambar yang akan dipakai
- maxc – Maximum number of corners we want [Negative values gives all the corners]
- Quality – Quality level parameter (preferred value=0.01)
- maxD – Maximum distance (preferred value=10)

````python

shitom = cv2.imread(path) 
gray_image = cv2.cvtColor(shitom, cv2.COLOR_BGR2GRAY) 
  
   
corners = cv2.goodFeaturesToTrack( 
    gray_image, maxCorners=25, qualityLevel=0.02, minDistance=10) 
corners = np.float32(corners) 
  
for item in corners: 
    x, y = item[0] 
    x = int(x) 
    y = int(y) 
    cv2.circle(shitom, (x, y), 6, (0, 255, 0), -1) 


plt.imshow(shitom)
plt.show()
````

# SIFT (Scale-Invariant Feature Transform)

````python
shift = cv2.imread(path) 
gray_image = cv2.cvtColor(shift, cv2.COLOR_BGR2GRAY) 
  

sift = cv2.SIFT_create() 
kp, des = sift.detectAndCompute(gray_image, None) 
  

kp_image = cv2.drawKeypoints(image, kp, None, color=( 
    0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS) 
plt.imshow(kp_image)
plt.show()
````

````python
fastim = cv2.imread(path) 
gray_images = cv2.cvtColor(fastim, cv2.COLOR_BGR2GRAY) 

fast = cv2.FastFeatureDetector_create() 
fast.setNonmaxSuppression(False) 

kps = fast.detect(gray_image, None) 
kp_images = cv2.drawKeypoints(fastim, kps, None, color=(0, 255, 0)) 

plt.imshow(kp_images)
plt.show()
````

# ORB (Oriented FAST and Rotated Brief)

````python
orbimage = cv2.imread(path) 
gray_imagess = cv2.cvtColor(orbimage, cv2.COLOR_BGR2GRAY)

orb = cv2.ORB_create(nfeatures=100) 
kpss, desss = orb.detectAndCompute(gray_imagess, None) 

kp_imagess = cv2.drawKeypoints(orbimage, kpss, None, color=(0, 255, 0), flags=0)

plt.imshow(kp_imagess)
plt.show()
````

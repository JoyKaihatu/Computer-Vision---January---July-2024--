```python
import cv2
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
import pandas as pd
```

```python
path = 'UTS_FILE/2-Pentagon.gif'
testing = cv2.VideoCapture(path)

ret, frame = testing.read()

print(frame.shape)

plt.imshow(frame)
```

```python
imgflt = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
imgflt = np.float32(imgflt)

dst = cv2.cornerHarris(imgflt, blockSize=2, ksize=3,k=0.6)

```

```python
dst = cv2.dilate(dst, None) 
frame[dst > 0.01 * dst.max()] = [0, 255, 0] 
```

```python
plt.imshow(frame)
```

```python
num_corner = np.sum(dst > 0.01 * dst.max())

print(num_corner)
```

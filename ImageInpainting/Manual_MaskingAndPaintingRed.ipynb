```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the damaged image
damaged_img = cv2.imread(filename=r"Scenic_View_Damage_Final.png")
 
# get the shape of the image
height, width = damaged_img.shape[0], damaged_img.shape[1]
 
# Converting all pixels greater than zero to black while black becomes white
for i in range(height):
    for j in range(width):
        if damaged_img[i, j].sum() == 255 and damaged_img[i,j][2] == 255:
            damaged_img[i, j] = [255,255,255]
        else:
            damaged_img[i, j] = [0,0,0]
 
# saving the mask 
mask = damaged_img
cv2.imwrite('Scenic_View_Damage_Final_Mask.png', mask)
 
# displaying mask
# cv2.imshow("damaged image mask", mask)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(mask)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
 
# Open the image.
img = cv2.imread('Scenic_View_Damage_2.png')
 
# Load the mask.
mask = cv2.imread('Scenic_View_Damage_2_Mask.png', 0)
 
# Inpaint.
dst = cv2.inpaint(img, mask, 10, cv2.INPAINT_NS)
 
# Write the output.
cv2.imwrite('Scenic_View_Damage_2_Rad10.png', dst)

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
